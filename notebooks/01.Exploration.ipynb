```python
from collections import Counter
from glob import glob
import json
import os
import pandas as pd
```

```python
HISTLOC = '../history/'
```

```python
!cat {HISTLOC}/2016-08-19.txt
```

```python
frases = []
for day in glob(os.path.join(HISTLOC, "*.txt")):
    with open(day, 'r') as f:
        for line in f:
            # Casos especiais
            if 'horário de brasília' in line.lower():
                continue
            
            # Decreto de qualificação do host interino!
            if 'qualificação do host interino' in line.lower():
                continue
            
            # Em geral:
            novas = [l.strip() for l in line.strip().split('.') if l]
            for frase in novas:
                separadas = frase.strip().split(',')
                if ',' in frase and len(separadas) > 3:
                    for s in separadas:
                        s = s.strip()
                        s = s[0].upper() + s[1:]
                        frases.append(s)
                elif 'cumpra' in frase.lower() or 'decreto' in frase.lower():
                    continue
                else:
                    frase = frase[0].upper() + frase[1:]
                    frases.append(frase)
```

```python
c = Counter(frases)
c.most_common()
```

```python
with open('../frases.json',  'wt', encoding='utf-8') as f:
    data = json.dumps(frases, ensure_ascii=False)
    f.write(data)
```
